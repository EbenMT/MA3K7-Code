```python
import random
import matplotlib.pyplot as plt

def simulate_process(n):

    hat = list(range(1, n+1))
    while len(hat) > 1:
        drawn = random.sample(hat, 2)
        diff = max(drawn) - min(drawn)
        for piece in drawn:
            hat.remove(piece)
        hat.append(diff)
    return hat[0]

def run_simulations(num_trials, n):
    outcomes = {}
    for _ in range(num_trials):
        result = simulate_process(n)
        outcomes[result] = outcomes.get(result, 0) + 1
    return outcomes

def plot_results(outcomes, num_trials, n):
    keys = sorted(outcomes.keys())
    counts = [outcomes[k] for k in keys]
    
    plt.figure(figsize=(8, 6))
    plt.bar(keys, counts, color='skyblue', edgecolor='black')
    plt.xlabel("Final Paper Value")
    plt.ylabel("Frequency")
    plt.title(f"Simulation of Final Paper Values (n = {n}) over {num_trials} Trials")
    plt.xticks(keys)
    plt.show()

if __name__ == "__main__":
    num_trials = 10000 
    n = 6                
    outcomes = run_simulations(num_trials, n)
    print("Outcomes:", outcomes)
    plot_results(outcomes, num_trials, n)
```
